```python
# Exploratory Data Analysis
```

```python
# Age distribution
This is a histogram that shows the distribution of patient ages. 
It can help understand the age group most affected by certain medical conditions.
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Gender distribution
This is a pie chart that shows the proportion of male to female patients. 
It can help understand the gender predominance in the dataset.
```

```python
plt.figure(figsize=(6,6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
# Blood Type distribution
This is a pie chart that shows the distribution of different blood types in the dataset. 
It can help understand the common blood types among the patients.
```

```python
plt.figure(figsize=(6,6))
df['Blood Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Blood Type Distribution')
plt.show()
```

```python
# Medical Condition distribution
This is a bar chart that shows the frequency of each medical condition. 
It can help identify the most common medical conditions in the dataset.
```

```python
plt.figure(figsize=(10,6))
df['Medical Condition'].value_counts().plot(kind='bar')
plt.title('Medical Condition Distribution')
plt.show()
```

```python
# Billing Amount distribution
This is a histogram that shows the distribution of the billing amounts. 
It can help understand the cost of healthcare.
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Billing Amount'], bins=30, kde=True)
plt.title('Billing Amount Distribution')
plt.show()
```

```python
# Admission Type distribution
This is a pie chart that shows the distribution of different types of admissions (emergency, routine, etc.). 
It can help understand the nature of most admissions.
```

```python
plt.figure(figsize=(6,6))
df['Admission Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Admission Type Distribution')
plt.show()
```
